````python
# Ejemplo_03: Se definirá un string que inicialmente comienza con comillas dobles y
# termina con comillas simples
print("hola mundo')
````

## Operadores y Operaciones sobre Strings

Al igual que con los tipos de datos numéricos, Python ofrece dos **operadores** para trabajar con los Strings.

Estos operadores son utilizados la realizar cálculos matemáticos sencillos y sus símbolos son **`+`** y **`*`**, los cuales representan a la suma y multiplicación de números respectivamente.

Al utilizar los símbolos **`+`** y **`*`** con strings, estos pasa a llamarse concatenación y repetición.

### Concatenación

Esta operación consiste en **unir** dos o más cadenas de `strings` para formar una nueva cadena
```
string_1 + string_2 -> string_1string_2
```

El **Ejemplo_08** muestra lo indicado anteriormente:

````python
# Ejemplo_08: concatenación de dos strings

texto_1 = "hola"
texto_2 = "mundo"

nuevo_texto = texto_1 + texto_2
nuevo_texto_2 = texto_1 + " " + texto_2

print(nuevo_texto)
print(nuevo_texto_2)
````

### Repetición

Esta operación se realiza entre una cadena de strings y un número entero. Consiste en **repetir** una cadena de `strings` tantas veces como indique el número al cuál este siendo multiplicado:

```
string_1 * 3 -> string_1string_1string_1
```

El **Ejemplo_09** muestra lo indicado anteriormente:

````python
# Ejemplo_09: repetición de un string

texto_1 = "hola"

nuevo_texto = texto_1 * 3
nuevo_texto_2 = 3 * texto_1

print(nuevo_texto)
print(nuevo_texto_2)
````

Además, al igual que en las listas, Python ofrece diversas operaciones que se pueden realizar con el tipo de dato `string`. Entre dichas operaciones se pueden destacar las siguientes funciones:
* **`len(<string>)`**: devuelve la cantidad de caracteres del string
* **`str(<expresion>)`**: devuelve la expresión ingresada en formato string

El **Ejemplo_10** ilustra lo mencionado anteriormente:

````python
# Ejemplo_10: uso de funciones de strings

oracion = "este es un ejemplo para calcular el largo"

largo = len(oracion)

numero = 66
texto_numero = str(numero)

numero_nuevo = numero*2
repeticion_numero = texto_numero*2

print(largo)
print(numero_nuevo)
print(repeticion_numero)
````

### Otras Operaciones

En el ejemplo anterior se calculó el largo o la cantidad de caracteres que componían el string **`oracion`**.

Posterior a ello, se transformó el número 66 a `string` y se realizó una comparación entre dos operaciones: multiplicación del número 66 por 2 y la multiplicación del número transformado a strings por 2 (repetición).

## Indexación

En esta misma línea de trabajo, existe una manera de acceder a los caracteres contenidos en el `string` de manera **individual**. 

Para lograr lo anterior, se tiene que llamar a la variable que contiene el `string` y acompañarlo de corchetes (`[ ]`) e indicar, dentro de el, la posición numérica (contando de izquierda a derecha) de dicho caracter.

No olvidar que al primer caracter le corresponderá el número `0` y así sucesivamente.

Tampoco olvidar que los espacios **también son considerado** al momento de contar los `strings`.

````python
# Ejemplo_11: Encontrar un caracter mediante su posición
palabra = "Hola Mundo"

# Se quiere mostar por pantalla la letra "M", por lo que tendremos que indicar su posición
caracter = palabra[5]

print(caracter)
````

Se debe tener presente que el primer caracter se encuentra en la posición **0** y el último caracter se encuentra en la posición **largo - 1** o, en otras palabras, **`len(<string>) - 1`**.

Dicho esto, se puede acceder a cualquier posición existente dentro del 0 y el largo - 1. Si se accede a una posición mayor a la anterior, el programa entregará un mensaje de error.

````python
# Ejemplo_12: encontrar un caracter mediante su posición

palabra = "Hola Mundo"

# Se quiere mostar por pantalla la letra "M", por lo que tendremos que indicar su posición
caracter = palabra[25]

print(caracter)
````

Además de acceder a posiciones determinadas, es posible acceder a determinadas zonas o partes de un `string`. Para ello, dentro del corchete se indicará 2 parámetros separados por dos puntos:

```
string[x:y]
```

*   **x**: corresponde a la primera posición que se quiere acceder.
*   **y**: corresponde a la posición posterior a la última que se quiere acceder (esta posición no es guardada).

````python
# Ejemplo_13: encontrar una cadena de caracteres mediante su posición

palabra = "Hola Mundo"

# Se accede a las posiciones 2 a 7 (no incluye la última)
cadena = palabra[2:8]

print(cadena)
````

## Métodos

Los strings, al igual que las listas, son **objetos**, que, en palabras simples, significa que poseen características y acciones propias. Esto último quiere decir que tienen acciones que les pertenecen, las cuales se denominan **métodos**. Por esta razón, los métodos de las listas no funcionan en los strings y viceversa.

Algunos de los métodos más utilizados son:

*   `<string>.lower()`: devuelve el mismo string pero en minúscula.
*   `<string>.upper()`: devuelve el mismo string pero en mayúscula.
*   `<string>.title()`: convierte la primera letra de cada palabra de una cadena en mayúscula.
*   `<string>.swapcase()`: convierte cada minúscula a mayúscula y viceversa.
*   `<string>.islower()`: devuelve `True` si los caracteres del strings son todas letras minúsculas.
*   `<string>.isupper()`: devuelve `True` si los caracteres del strings son todas letras mayúsculas.
*   `<string>.isdigit()`: devuelve `True` si los caracteres del strings son todos numéricos.
*   `<string>.isalpha()`: devuelve `True` si los caracteres del strings son todos letras.

````python
# Ejemplo_14: método .lower() y .upper()

palabra = "Hola Mundo"

palabra_minuscula = palabra.lower()
palabra_mayuscula = palabra.upper()

print(palabra_minuscula)
print(palabra_mayuscula)
````

````python
# Ejemplo_15: método .title() y .swapcase()

palabra = "hola mundo"

palabra_titulo = palabra.title()
palabra_cambiada = palabra_titulo.swapcase()

print(palabra_titulo)
print(palabra_cambiada)
````

````python
# Ejemplo_16: Método .islower() e isupper()

palabra = "Hola Mundo"
palabra_2 = "HOLA MUNDO"
palabra_3 = "hola mundo"

palabra_minuscula = palabra.islower()
palabra_minuscula_2 = palabra_3.islower()

palabra_mayuscula = palabra.isupper()
palabra_mayuscula_2 = palabra_2.isupper()

print(palabra_minuscula)
print(palabra_minuscula_2)

print(palabra_mayuscula)
print(palabra_mayuscula_2)
````

````python
#Ejemplo_17: Método is.alpha() y is.digit()

palabra = "HolaMundo"
palabra_2 = "espero que pronto sea 31 de diciembre"
palabra_3 = "15842"

palabra_letra = palabra.isalpha()
palabra_letra_2 = palabra_2.isalpha()

palabra_numero = palabra_2.isdigit()
palabra_numero_2 = palabra_3.isdigit()

print(palabra_letra)
print(palabra_letra_2)

print(palabra_numero)
print(palabra_numero_2)
````

En otras palabras, en el momento en que se crea un string, este no puede ser modificado, solo sobreescrito.

## Conversión de Tipo de Datos `list - str`

Al iniciarse en la programación con Python, es muy común trabajar con strings y listas, por lo que hay que saber cómo poder pasar de un tipo de dato a otro en cualquier momento.

### De strings a listas

Para convertir un `string` a `list`, se presentan 3 opciones:

1. Utilizar iteración (ciclo while).
2. Utilizar directamente la función de cambio de tipo `list()`.
3. Utilizar el método `<string>.split()`.

#### 1. Iteraciones

Se accederá a cada posicón de un determinado strings y se irá almacenando en una lista

````python
# Ejemplo_20: Uso de iteraciones para  crear una lista a partir de un string

texto = "hola mundo"
lista_texto = []

i = 0
while i < len(texto):
  lista_texto.append(texto[i])
  i = i + 1
print(lista_texto)
````

#### 2. Función `list()`

Esta función convierte cualquier dato iterable a una lista.
*   Si se invoca sin parámetros de entrada, generará una lista vacía.
*   Si se invoca con un string, cada elemento de la lista tendrá un caracter de dicho string.

````python
# Ejemplo_21: Uso de la función list() para  crear una lista

texto = "hola mundo"
lista_texto = list(texto)

print(lista_texto)
````

#### 3. Método `<string>.split()`

Este método pertenece a los strings y permite dividir una determinada cadena de caracteres por un determinado "substring", formando una lista. Dicho substring es otro string que corresponde al parámetro del método.

````python
# Ejemplo_22a: uso del método .split() para  crear una lista

texto = "Este es un ejemplo para que se separe en palabras"

# En este caso se quiere separar las palabras, por lo que el caracter que se debe considerar
# para separar el string es un espacio, " "
lista_texto = texto.split(" ")

print(lista_texto)
````

En este caso, el string contiene números separados por punto y coma, más un espacio:

````python
# Ejemplo 22b: uso de split con un substring
texto = "1; 2; 3; 4"
# Separamos por "; ", que es lo que distingue a los elementos entre sí
lista_texto = texto.split("; ")

print(lista_texto)
````

### De listas a strings

Del mismo modo, como existen opciones para poder transformar un string a lista, también existen alternativas para realizar el proceso inverso. Para ello, se presentan 2 opciones:
1.   Uso de la función `str()`
2.   Uso del método `<string>.join(lista)`

#### 1. Función `str()`

Esta función convierte los tipos de datos que ingresan a la función a string. Esto puede provocar inconvenientes al momento de transformar una lista a string, puesto que las llaves `[ ]` también son convertidas: Python transforma la **representación** de la lista a string (como lo que muestra cuando pasamos la lista a `print`):

````python
# Ejemplo_23: uso de la función `str()` para  crear un string

lista_texto = ["hola", " ", "mundo"]

#En este caso se quiere transformar la lista a string con la función str()
texto = str(lista_texto)

print(texto)
````

#### 2. Método `<string>.join(lista)`

Este método pertenece a los strings y permite agrupar los elementos de una determinada lista de strings en una cadena de caracteres, unidos por el string que ejecuta el método.

````python
# Ejemplo_24: uso del método .split() para crear una lista

lista_texto = ['Este', 'es', 'un', 'ejemplo', 'para', 'que', 'se', 'una',
               'una', 'cadena', 'de', 'texto']

# En este caso, se quiere unir los elementos de una lista, por lo que el caracter que se debe
# considerar para unir los elementos es un espacio, " "
texto = " ".join(lista_texto)

print(texto)
````
