# Algoritmo

## Definición

Existen diversas definiciones para dar a conocer lo que se entiende por algoritmo.
Dependiendo del área de estudio en la que se utilice, se dará mayor énfasis a unas cosas que a otras.

No obstante, para simplificar su entendimiento se puede decir que:

*Un **algoritmo** es una lista definida, ordenada y finita de pasos/procesos que permiten hallar una solución a un determinado problema*

## Características

Los Algoritmos, para ser considerados  efectivos, deben poseer las siguientes características:

1. Deben tener  cero o más datos de **entrada**.
2. Debe producir un resultado. Los datos de **salida** serán los resultados al efectuar una serie de instrucciones.
3. Es **finito**, siempre debe terminar posterior a un número determinado de pasos.
4. Cada uno de los pasos declarados deben estar definidos con **precisión**, sin generar ambigüedades.
5. Debe ser **definido**. Si se sigue un algoritmo dos veces, se debe obtener el mismo resultado.

Se debe ser consciente respecto a que un algorimo es el cimiento de cualquier solución, independiente el área o el contexto. No es un concepto que solo se aplique en la programación, sino que está inmerso en todo lugar en que se tenga un problema y se formule o elabore una solución.

## Clasificación

Según el tipo de problema al cual se enfrentan, la solución propuesta mediante un algoritmo puede clasificarse en 2 grandes grupos:

*   **Algoritmos Cuantitativos**: soluciones propuesta a problemáticas en donde intervienen números de manera directa:
    *   Cálculo de una sumatoria.
    *   Obtención del despeje de una ecuación.
*   **Algoritmos Cualitativos**: soluciones propuestas a problemáticas donde no hay números involucrados. Son situaciones cotidianas:
    *   Cómo iniciar sesión en una página web.
    *   Cómo instalar un juego o programa computacional.
    *   Cómo cocinar un postre.

## Representación

Teniendo clara la solución a una determinada problemática, el algoritmo puede presentarse de diversas maneras. Dependerá tanto de la persona, como del problema y si existen más usuarios involucrados.

Entre las representaciones más utilizadas se encuentran:

*  **Lista de cotejo** - Recetas de cocina o instrucciones básicas
*  **Fórmulas matemáticas** - Resolución de problemas numéricos
*  **Diagramas de proceso/flujos** - Propuestas de trabajo u organizacionales
*  **Pictóricos** - Explicaciones personales o para infantes

## Ejercicio

Para comprobar la importancia de la generación de un buen algoritmo, se plantea la siguiente actividad:

Describa el proceso, paso a paso, de la preparación de un huevo revuelto. Para ello, considere que usted se encuentra en la cocina y en ella están todos los implementos y materiales necesarios.

Una vez realizado el proceso, intercámbielo con otro/a estudiante y comenten lo sucedido.
