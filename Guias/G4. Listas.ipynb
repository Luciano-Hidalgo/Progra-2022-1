## PedirListas.py

````python
# MAGIA NEGRA QUE PIDE UNA LISTA, Y LA TRANSFORMA AUTOM√ÅTICAMENTE A UNA LISTA DE ENTEROS
lista = list(map(int,input("Ingrese una lista de n√∫meros enteros, separados por espacio:").split(" ")))

# Se imprime la lista para verificar que tenga lo que queremos
print(lista)
````

````python
##########################################################################################
# EXPLICACI√ìN
##########################################################################################
# PASO 1: Se solicita una entrada usando input(), esta ser√° un string
lista = input("Ingrese una lista de n√∫meros enteros, separados por espacio:")

# PRINTS PARA MOSTRAR EL PASO
print('PASO 1:')
print('\t', lista)

# PASO 2: Se utiliza el m√©todo .split(" ") para cortar el string por el caracter espacio
# obteniendo as√≠ una lista de strings
lista = lista.split(" ")

# PRINTS PARA MOSTRAR EL PASO
print('PASO 2:')
print('\t', lista)

# PASO 3: Se utiliza la funci√≥n map(), lo que map hace es:
# Aplica la funci√≥n entregada en el primer par√°metro (en este caso int)
# Al objeto del segundo par√°metro (en este caso la lista de string)
# Devolviendo un objeto de tipo map, el cu√°l es √∫til para iterar pero no para
# almacenar o modificar valores
lista = map(int, lista)

# PRINTS PARA MOSTRAR EL PASO
print('PASO 3:')
print('\t', lista)
# Como vemos, map nos entrega un mensaje de que algo existe en memoria, pero no muestra los valores

# PASO 4: Se realiza la conversi√≥n de map a lista, lo que nos permite volver a aplicar
# todas las operaciones de listas sobre la entrada
lista = list(lista)

# PRINTS PARA MOSTRAR EL PASO
print('PASO 4:')
print('\t', lista)
````

### Otras implementaciones de la misma soluci√≥n.

````python
# MAGIA NEGRA QUE PIDE UNA LISTA, Y LA TRANSFORMA AUTOM√ÅTICAMENTE A UNA LISTA DE ENTEROS
lista = list(eval(input("Ingrese una lista de n√∫meros enteros, separados por coma:"))

# Se imprime la lista para verificar que tenga lo que queremos
print(lista)

##########################################################################################
# EXPLICACI√ìN
##########################################################################################
# PASO 1: Se solicita una entrada usando input(), esta ser√° un string
lista = input("Ingrese una lista de n√∫meros enteros, separados por coma:")

# PRINTS PARA MOSTRAR EL PASO
print('PASO 1:')
print('\t', lista)

# PASO 2: Se utiliza el m√©todo eval para transformar el string en una expresi√≥n en Python
lista = eval(lista)

# PRINTS PARA MOSTRAR EL PASO
# En este caso, el tipo de dato almacenado es una tupla, lo que es parecido a una lista
# pero es inmutable
print('PASO 2:')
print('\t', lista)

# PASO 3: Se utiliza la funci√≥n list() para transformar la tupla en lista
lista = list(lista)

# PRINTS PARA MOSTRAR EL PASO
print('PASO 3:')
print('\t', lista)
````

````python
# MAGIA NEGRA QUE PIDE UNA LISTA, Y LA TRANSFORMA AUTOM√ÅTICAMENTE A UNA LISTA DE ENTEROS
lista = eval('['+ ','.join(input('Ingrese una lista de enteros, separados por espacio: ').split(' ')) + ']')

# Se imprime la lista para verificar que tenga lo que queremos
print(lista)

##########################################################################################
# EXPLICACI√ìN
##########################################################################################
# PASO 1: Se solicita una entrada usando input(), esta ser√° un string
lista = input("Ingrese una lista de n√∫meros enteros, separados por espacio:")

# PRINTS PARA MOSTRAR EL PASO
print('PASO 1:')
print('\t', lista)

# PASO 2: Se utiliza el m√©todo split para transformar el string en una lista
lista = lista.split(' ')

# PRINTS PARA MOSTRAR EL PASO
print('PASO 2:')
print('\t', lista)

# PASO 3: Se utiliza el m√©todo .join() para juntar todos los elementos de la lista en un string separado por comas
# N√≥tese que el join recibe la lista como argumento y el separador como objeto
lista = ','.join(lista)

# PRINTS PARA MOSTRAR EL PASO
print('PASO 3:')
print('\t', lista)

# PASO 4: Se concatenan los par√©ntesis al string a fin de formar una lista en Python
lista = '[' + lista + ']'

# PRINTS PARA MOSTRAR EL PASO
print('PASO 4:')
print('\t', lista)

# PASO 5: Se utiliza la funci√≥n eval() para transformar el string en una expresi√≥n v√°lida en Python, en este caso, una lista
lista = eval(lista)

# PRINTS PARA MOSTRAR EL PASO
print('PASO 5:')
print('\t', lista)
````

## GeneradorListas.py

````python
# Generador de n√∫meros enteros
from random import randint

# ENTRADA
# Solicito la cantidad de valores para la lista
cantidad = int(input("Ingrese la cantidad de valores que desea: "))

# Solicito el m√≠nimo del rango
minimo = int(input("Ingrese el valor m√°s bajo del rango: "))

# Solicito el m√°ximo del rango
maximo = int(input("Ingrese el valor m√°s alto del rango: "))

# PROCESO
# Declaro la lista vacia
lista = []

# Mientras la lista no tenga la cantidad de elementos
while len(lista) < cantidad :
    #  Se genera un n√∫mero aleatorio
    numero = randint(minimo, maximo)
    # Se agrega el numero a la lista
    lista.append(numero)
# SALIDA
# Se imprime la lista
print(lista)
````
