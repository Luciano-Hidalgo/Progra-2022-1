# Objetos

## Definición

Los objetos son una manera de organizar los tipos de datos y relacionarlos para poder manejarlos de manera más sencilla. En otras palabras, los objetos son una forma ordenada de agrupar datos y realizar determinadas operaciones sobre ellos (métodos).

La finalidad de los objetos es representar algo real o existente en un lenguaje de programación. Por ello, un objeto posee datos o atributos y métodos o funciones.


## Tipos de Objetos

Desde que se inició este curso introductorio, se han utilizado objetos sin mencionarlos, puesto que  todos los tipos de datos que provee Python son objetos.

Los tipos de datos que se manejan en Python tienen características en común:

* Listas: son un tipo de dato que se utiliza para almacenar valores (como una especie de caja) y como tal, tienen sus propias características y sus propios métodos, los cuales no pueden aplicarse a otros tipos de datos (agregar, eliminar, insertar...).

* Strings: son un tipo de dato que se utiliza para trabajar con cadenas de texto y como tal, tienen sus propias características y sus propios métodos, los cuales no pueden aplicarse a otros tipos de datos (transformar caracteres a mayúscula, a minúscula, capitalizar texto, preguntar si está compuesto solo por letras, solo por número, combinación de ambas...).

Se mencionan estos 2 tipos de datos, puesto que son los que más se utilizarán durate el desarrollo de este curso.

Más adelante, cuando se vean los tópicos de `strings` y `listas`, se profundizarán más las características y los distintos métodos que se pueden utilizar en cada uno de estos tipos de datos.

A modo de complementar el tema de objetos, en Python se pueden encontrar tipos de datos llamados `tuplas` y `diccionario`, los cuales no son parte del programa pero se mencionan por la relevancia que estos tienen.

Si bien, Python provee un gran número de objetos definidos, en muchas situaciones esta cantidad será insuficiente para resolver problemas más complejos, por ello, Python ofrece la posibilidad de crear nuevos objetos, sin embargo, esto está más allá del alcance de este curso introductorio.

## Atributos y Métodos

Los **atributos** son las características o información propia que tendrán los objetos.

Por otra parte, los **métodos** son las acciones que este objeto realiza en el entorno.

A modo de ejemplo, para facilitar los conceptos vistos, se plantea la siguiente situación:
* Se tiene un **objeto** del mundo real - un auto negro
* Sus **atributos** pueden ser - Color, cantidad de puertas, velocidad que alcanza...
* Sus **métodos** podrían ser - encendido de motor, acelaración, frenado...

Considerando un segundo ejemplo:
* **Objeto**: taza de café
* **Atributos**: color, forma, tamaño, material
* **Métodos**: sostenerse, almacenar
