## Operadores de comparaci√≥n

Para poder realizar comparaciones entre 2 expresiones aritm√©ticas se deben utilizar los operadores de comparaci√≥n. Python provee un listado de los operadores de comparaci√≥n m√°s utilizados:

Cabe se√±alar que el resultado de utilizar estos operadores genera el tipo de dato l√≥gico o `booleano` (**`bool`**).

|Operaci√≥n     |Operador |Ejemplo       |Resultado   |
|:------------:|:-------:|:------------:|:----------:|
|Mayor         | `>`     | 2.0 > 5      | **False**  |
|Mayor o igual | `>=`    | 8 >= 3.7     | **True**   |
|Menor         | `<`     | 1.5 < 1.4    | **False**  |

|Igual         | `==`    | 2.0 == -2    | **False**  |
|Distinto      | `!=`    | 4.0 != 4.01  | **True**   |

```python
expresion_2 = x > 5
print(expresion_2)
```

```python
expresion_3 = x < 10
print(expresion_3)
```

```python
expresion_4 = x >= 5
print(expresion_4)
```

```python
expresion_5 = x<= 10
print(expresion_5)
```

## Operadores l√≥gicos

Para poder agrupar operaciones l√≥gicas/booleanas, se utilizan los operadores l√≥gicos. Para ello, Python provee tres operadores b√°sicos, los cuales se muestran a continuaci√≥n:

*   Conjunci√≥n o *y l√≥gico* (**and**)
*   Disyunci√≥n u *o l√≥gico* (**or**)
*   Negaci√≥n (**not**)

Cada operador l√≥gico trabaja con un conjunto de reglas resumidas en tablas de verdad. 

La tabla de verdad ilustra el resultado de operar con 1 o 2 valores l√≥gicos, seg√∫n corresponda.


### Conjunci√≥n o **y l√≥gico**

El operador conjunci√≥n (**`and`**) es un operador binario (necesita dos expresiones para trabajar). Puede generar dos resultados, dependiendo de las expresiones evaluadas:

* Genera un valor verdadero (**`True`**) si y solo si ambas expresiones evaluadas son verdaderas (`True` **and** `True`)
* Genera un valor falso (**`False`**) si por lo menos alguna de las expresiones evaluadas es falsa (**`False`**)

|expresi√≥n_1  |expresi√≥n_2  |expresion_1 and expresion_2|
|:-----------:|:-----------:|:-------------------------:|
|True         |True         | **True**                  |
|True         |False        | **False**                 |
|False        |True         | **False**                 |
|False        |False        | **False**                 |

```python
expresion_1 = 4.8 >= 4.2
expresion_2 = 2 == 2

evaluacion_1 = expresion_1 and expresion_2

print(evaluacion_1)
```

```python
expresion_1 = 4.8 >= 4.2
expresion_3 = 3 > 5

evaluacion_2 = expresion_1 and expresion_3

print(evaluacion_2)
```

```python
expresion_2 = 2 == 2
expresion_3 = 3 > 5

evaluacion_3 = expresion_3 and expresion_2

print(evaluacion_3)
```

```python
expresion_3 = 3 > 5
expresion_4 = 91 <= 90

evaluacion_4 = expresion_3 and expresion_4

print(evaluacion_4)
```

### Disyunci√≥n u **o l√≥gico**

El operador de disyunci√≥n (**`or`**) es un operador binario (necesita dos expresiones para trabajar). Puede generar dos resultados, dependiendo de las expresiones evaluadas:

* Genera un valor verdadero (**`True`**) si por lo menos alguna de las expresiones evaluadas es verdadera (**`True`**)
* Genera un valor falso (**`False`**) si y solo si ambas expresiones evaluadas son falsas (`False` **or** `False`)

|expresi√≥n_1  |expresi√≥n_2  |expresion_1 or expresion_2|
|:-----------:|:-----------:|:------------------------:|
|True         |True         | **True**                 |
|True         |False        | **True**                 |
|False        |True         | **True**                 |
|False        |False        | **False**                |

```python
expresion_1 = 4.8 >= 4.2
expresion_2 = 2 == 2

evaluacion_1 = expresion_1 or expresion_2

print(evaluacion_1)
```

```python
expresion_1 = 4.8 >= 4.2
expresion_3 = 3 > 5

evaluacion_2 = expresion_1 or expresion_3

print(evaluacion_2)
```

```python
expresion_2 = 2 == 2
expresion_3 = 3 > 5

evaluacion_3 = expresion_3 or expresion_2

print(evaluacion_3)
```

```python
expresion_3 = 3 > 5
expresion_4 = 91 <= 90

evaluacion_4 = expresion_3 or expresion_4

print(evaluacion_4)
```

### Negaci√≥n

El operador de negaci√≥n (**`not`**) es un operador unario (trabaja solo con una expresi√≥n).
Puede generar dos resultados, dependiendo la expresi√≥n evaluada:

* Genera un valor verdadero (**`True`**) si y solo si la expresi√≥n evaluada es falsa (**`False`**)
* Genera un valor falso (**`False`**) si y solo si la expresi√≥n evaluada es verdadera (**`True`**)

|expresi√≥n_1  |not expresion_1 |
|:-----------:|:--------------:|
|True         | **False**      |
|False        | **True**       |

```python
expresi√≥n_1 = 3 < 5
evaluacion_1 = not expresion_1

print(evaluacion_1)
```

```python
expresi√≥n_2 = 45 >= 45.1
evaluacion_2 = not expresion_2

print(evaluacion_2)
```

## Expresiones booleanas

Al igual que los operadores aritm√©ticos, los operadores l√≥gicos tienen su propia **precedencia establecida**:

1.   Operadores de comparaci√≥n
2.   Negaci√≥n (**`not`**)
3.   Conjunciones (**`and`**)
4.   Disyunciones (**`or`**)

Cabe se√±alar que si una expresi√≥n contiene operaciones aritm√©ticas, √©stas son resueltas antes de utilizar los operadores de comparaci√≥n (**precedencia de operadores**)

Realice los siguientes ejercicios y compare con los resultados entregados por Python. Para ello, considere que la variable **`x`** ser√° evaluada primero con el valor 5 y luego, con el valor 10.

Ejercicio.PNG

```python
# Se solicitar√° el valor de la variable x al usuario 
x = input("Favor ingresar el valor de la variable x: ")

# Se realiza el cambio de tipo de dato a entero
x = int(x)

# Se evalua la expresi√≥n booleana
expresion_1 = x + 1 > x**2 or x + 5 <= 2 * x

# Se muestra por pantalla el resultado obtenido
print(expresion_1)
```

```python
# Se solicitar√° el valor de la variable x al usuario 
x = input("Favor ingresar el valor de la variable x: ")

# Se realiza el cambio de tipo de dato a entero
x = int(x)

# Se evalua la expresi√≥n booleana
expresion_2 = x < 6 and not x >= 10 or not 2**x < 16

# Se muestra por pantalla el resultado obtenido
print(expresion_2)
```

```python
# Se solicitar√° el valor de la variable x al usuario 
x = input("Favor ingresar el valor de la variable x: ")

# Se realiza el cambio de tipo de dato a entero
x = int(x)

# Se evalua la expresi√≥n booleana
expresion_3 = not(x > 3 and x<=10) and (x**2 <= 16 or 2 * x <= x + 10)

# Se muestra por pantalla el resultado obtenido
print(expresion_3)
```

La utilidad de las expresiones booleanas, en programaci√≥n, es la de entregar la posibilidad a un programa de **tomar decisiones**

## Sentencias condicionales

Ahora que ya se han dado a conocer los operadores de comparaci√≥n y operadores l√≥gicos, es posible indicarle a los programas c√≥mo trabajar en distintas situaciones.

Para poder indicarle a Python sobre qu√© camino seguir en la ejecuci√≥n de un c√≥digo, se utilizar√°n las sentencias **`if`** e **`if-else`**, las cuales ser√°n definidas a continuaci√≥n:

### Sentencia if

Los lenguajes de programaci√≥n proveen distintas **estructuras de decisi√≥n**. La m√°s simple se compone de **una condici√≥n**, que no es otra cosa que una expresi√≥n booleana, y un bloque de sentencias condicionales que solo se ejecutar√° si la condici√≥n se eval√∫a con un valor verdadero (**`True`**).

El siguiente diagrama de flujo ilustra la ejecuci√≥n de una estructura de decisi√≥n simple.

if.PNG

La **sintaxis** (forma de escribir) que Python utiliza para una estructura de decisi√≥n simple es la siguiente:

```python
if <condici√≥n>:
  # Se ejecuta si la condici√≥n se cumple
  <Bloque de sentencias condicionales>
# Se ejecuta independiente de si al condici√≥n se cumple o no
<Bloque de sentencias que sigue>
```

A continuaci√≥n, se da un ejemplo donde se entregar√° un mensaje al usuario si el n√∫mero que se ingresa por teclado es par:

```python
# ENTRADA
# Se solicita el n√∫mero al usuario
numero = input("Favor ingresar n√∫mero para evaluar su paridad: ")

# PROCESAMIENTO
# Se realiza el cambio de tipo de dato a n√∫mero entero
numero = int(numero)

# Se eval√∫a la paridad del n√∫mero ingresado
if numero%2 == 0:
  print("El n√∫mero ingresado es par")
print("Fin del programa")
```

### Sentencia if-else

Una alternativa muy utilizada al momento de trabajar con sentencias condicionales es el uso de decisiones alternativas. Esta se compone de una expresi√≥n booleana, y un bloque de sentencias condicionales que solo se ejecutar√° si la condici√≥n se eval√∫a con un valor verdadero (**`True`**). En caso de que la condici√≥n evaluada arroje un valor falso (**`False`**), se ejecutar√° la condici√≥n que sucede a la palabra reservada **`else`**.

El siguiente diagrama de flujo ilustra la ejecuci√≥n de una estructura de decisi√≥n simple.

if-else.PNG

La **sintaxis** (forma de escribir) que Python utiliza para una estructura de decisi√≥n alternativa es la siguiente:

```python
if <condici√≥n>:
  # Se ejecuta si la condici√≥n se cumple
  <Bloque de sentencias condicionales>
else:
  # Se ejecuta si la condici√≥n NO se cumple
  <Bloque de sentencias alternativo>
# Se ejecuta independiente de si al condici√≥n se cumple o no
<Bloque de sentencias que sigue>
```

A continuaci√≥n se utilizar√° el ejemplo anterior, donde se entregar√° un mensaje al usuario si el n√∫mero que se ingresa por teclado es par o, en su defecto, es impar:

```python
# ENTRADA
# Se solicita el n√∫mero al usuario
numero = input("Favor ingresar n√∫mero para evaluar su paridad: ")

# PROCESAMIENTO
# Se realiza el cambio de tipo de dato a n√∫mero entero
numero = int(numero)

# Se eval√∫a la paridad del n√∫mero ingresado
if numero%2 == 0:
  print("El n√∫mero ingresado es par")
else:
  print("El n√∫mero ingresado NO es par")
print("Fin del programa")
```

### Decisiones consecutivas y anidadas

Antes de terminar, se propone el siguiente ejercicio para practicar los conceptos vistos:

Dise√±e un programa que solicite por teclado la edad de 2 personas e indique qui√©n es el m√°s joven. Tener en consideraci√≥n que ambas personas pueden tener la misma edad, por lo que se debe generar un mensaje adecuado para cada caso

```python
# ENTRADA
# Se solicita la edad de las personas respectivas
persona_1 = input("Favor ingresar la edad de la primera persona: ")
persona_2 = input("Favor ingresar la edad de la segunda persona: ")

# PROCESAMIENTO
# Se realiza el cambio de tipo de dato, en este caso se debe cambiar a n√∫mero entero
persona_1 = int(persona_1)
persona_2 = int(persona_2)

# Se realizan las sentencias condicionales que se solicitan en el enunciado
if persona_1 < persona_2:
  print("La primera persona es menor")
if persona_1 > persona_2:
  print("La segunda persona es menor")
if persona_1 == persona_2:
  print("Ambas personas tienen la misma edad")
```

En el ejercicio anterior, se puede ver como se trabaj√≥ con **m√∫ltiples decisiones consecutivas**, donde en cada caso se especifica la condici√≥n respectiva.

Ahora se presenta el mismo ejemplo pero considerando para la soluci√≥n el uso de **decisiones anidadas**.

```python
# ENTRADA
# Se solicita la edad de las personas respectivas
persona_1 = input("Favor ingresar la edad de la primera persona: ")
persona_2 = input("Favor ingresar la edad de la segunda persona: ")

# PROCESAMIENTO
# Se realiza el cambio de tipo de dato, en este caso se debe cambiar a n√∫mero entero
persona_1 = int(persona_1)
persona_2 = int(persona_2)

# Se realizan las sentencias condicionales que se solicitan en el enunciado
if persona_1 < persona_2:
  print("La primera persona es menor")
else:
  if persona_1 > persona_2:
    print("La segunda persona es menor")
  else:
    print("Ambas personas tienen la misma edad")
```

Al ver la resoluci√≥n del ejercicio anterior se puede apreciar el uso de **`if-else`**, pero dentro de la condici√≥n **`else`** existen m√°s condiciones. Lo anterior se conoce como **decisiones anidadas**.

Cabe se√±alar que el **`else`** es la negaci√≥n absoluta del **`if`** que se encuentra **inmediatamante superior a este**.

Para dejar m√°s clara la idea anterior, se desarrollar√° nuevamente el ejercicio anterior pero usando decisiones consecutivas y **`else`**:

```python
#ENTRADA
#Se solicita la edad de las personas respectivas
persona_1 = input("Favor ingresar la edad de la primera persona: ")
persona_2 = input("Favor ingresar la edad de la segunda persona: ")

#PROCESAMIENTO
#Se realiza el cambio de tipo de dato, en este caso se debe cambiar a n√∫mero entero
persona_1 = int(persona_1)
persona_2 = int(persona_2)

#Se realizan las sentencias condicionales que se solicitan en el enunciado
if persona_1 < persona_2:
  print("La primera persona es menor")
if persona_1 > persona_2:
  print("La segunda persona es menor")
else:
  print("Ambas personas tienen la misma edad")
```

Como se puede observar, si la primera decisi√≥n se cumple (si la primera persona es menor) se entrega el mensaje por pantalla y como la segunda decisi√≥n no puede cumplirse (si la segunda persona es menor), se entrega el mesaje que conlleva el **`else`**. Esto se debe a lo mencionado anteriormente:

**El `else` es la negaci√≥n absoluta del `if` que se encuentra inmediatamante superior a este.**

Para reforzar los contenidos y aplicar las √∫ltimas indicaciones entregadas se propone el siguiente ejercicio:

En Chile, si las personas que tienen menos de 18 a√±os, son consideradas menores de edad; aquellas que est√°n en el rango etario de 18 a 35 a√±os se les considera adultos-j√≥venes; quienes superan los 35 a√±os, pero a√∫n no alzanzan los 65 a√±os, son consideradas adultos, y los que tiene 65 a√±os o m√°s son consideradas adultos mayores o miembros de la "tercera edad".

Se solicita que construya un programa en Python que solicite por teclado la edad de una persona e indique su categor√≠a a partir del rango etario dado.

Resoluci√≥n 1: **Decisiones consecutivas**

```python
# ENTRADA
# Se solicita la edad a la persona
edad_persona = input("Favor ingresar la edad de la persona en estudio: ")

# PROCESAMIENTO
# Se realiza el cambio de tipo de dato, en este caso en particular se transformar√° a n√∫mero entero
edad_persona = int(edad_persona)

# Se plantean las condiciones de acuerdo a lo estipulado en el enunciado
if edad_persona < 18:
  print("La persona es menor de edad")
if edad_persona >= 18 and edad_persona <= 35:
  print("La persona es adulto-joven")
if edad_persona > 35 and edad_persona < 65:
  print("La persona es adulto")
if edad_persona >= 65:
  print("La persona es adulto mayor")
```

Resoluci√≥n 2: **Decisiones anidadas**

```python
# ENTRADA
# Se solicita la edad a la persona
edad_persona = input("Favor ingresar la edad de la persona en estudio: ")

# PROCESAMIENTO
# Se realiza el cambio de tipo de dato, en este caso en particular se transformar√° a n√∫mero entero
edad_persona = int(edad_persona)

# Se plantean las condiciones de acuerdo a lo estipulado en el enunciado
if edad_persona < 18:
  print("La persona es menor de edad")
else:
  if edad_persona >= 18 and edad_persona <= 35:
    print("La persona es adulto-joven")
  else:
    if edad_persona > 35 and edad_persona < 65:
      print("La persona es adulto")
    else:
      print("La persona es adulto mayor")
```

### Indentaci√≥n

A diferencia de otros lenguajes de programaci√≥n, Python no utiliza s√≠mbolos para se√±alar el comienzo y el fin de un bloque de sentencias (if-else). En su lugar, Python utiliza la indentaci√≥n (sangr√≠a).

La importancia de la indentaci√≥n es separar el bloque de sentencias condicionales a ejecutar en caso de cumplir la condici√≥n dada. Su aparici√≥n se genera posterior a la utilizaci√≥n del caracter dos puntos (**`:`**).

El bloque de sentencias condicionales termina cuando la indentaci√≥n est√° al mismo nivel que la condici√≥n condicional que la origin√≥.
